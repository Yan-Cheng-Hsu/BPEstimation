```python
L = [ 'hr' ,
       'AS' ,  'DS' ,  'S1' , 'S2' , 'AA' , 'dAA' , 'sdAA' , 
       'DA' , 'dDA' , 'sdDA' , 'PI' , 'dPI' , 'sdPI' , 'dVI', 'sdVI',
       'AID' , 'dAID', 'sdAID' , 'dDID' , 'sdDID' , 'PIR' , 'sdRIPV',
       'AT', 'DT' ,
       'Slope_a' , 'S3' , 'S4' ,  'NI', 'AI' , 'RSD' , 'RSC']
Att = {}
```

```python
for index in range( len(Features[0][0])  ):
    Att[ L[index] ] = []
    for i in range( len(Features) ):
        for j in range( len(Features[i]) ):
            if Features[i][j][index] != 0:
                Att[ L[index] ].append( Features[i][j][index] )
    Att[ L[index] ] = np.array( Att[ L[index] ] ) 
for i in L:
        Att[ i ] = ( Att[ i ] - np.mean( Att[ i ] ) ) / np.std( Att[ i ] )
        Att[ i ] = np.round( Att[ i ] , 1 )
        Att[ i ] = np.sort( Att[ i ] , kind = 'heapsort' )
```
