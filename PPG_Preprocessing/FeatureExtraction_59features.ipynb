```python
for index in range( len(Features[0][0]) + 1  ):
    if index != 59:
        Att[ L[index] ] = []
        for i in range( len(Features) ):
            for j in range( len(Features[i]) ):
                if Features[i][j][index] != 0:
                    Att[ L[index] ].append( Features[i][j][index] )
        Att[ L[index] ] = np.array( Att[ L[index] ] ) 
    else:
        Att[ L[index] ] = []
        for i in range( len(Features) ):
            for j in range( len(Features[i]) ):
                if Features[i][j][53] != 0:
                    Att[ L[index] ].append( Features[i][j][53] )
        Att[ L[index] ] = np.array( Att[ L[index] ] ) 
        
        
for i in L:
    if i != 'ni':
        Att[ i ] = ( Att[ i ] - np.mean( Att[ i ] ) ) / np.std( Att[ i ] )
        Att[ i ] = np.round( Att[ i ] , 1 )
        Att[ i ] = np.sort( Att[ i ] , kind = 'heapsort' )
    else:
        Att[ i ] = np.sort( Att[ i ] , kind = 'heapsort' )
```

```python
import math
x = {}
y = {}
g = {}
for index in L:
    if index == 'ni':
        continue
    else:
        x[ index ] = []
        y[ index ] = []
        x[ index ].append( Att[ index ][0] )
        y[ index ].append( 1 )
        j = 0
        for i in range( 1 , len(Att[ index ]) , 1 ):
            if Att[ index ][i] == x[ index ][j]:
                y[ index ][j] = y[ index ][j] + 1
            else:
                x[ index ].append( Att[ index ][i] )
                y[ index ].append( 1 )
                j = j + 1
```

```python
for index in M:
    x[ index ] = []
    y[ index ] = []
    x[ index ].append( Att1[ index ][0] )
    y[ index ].append( 1 )
    j = 0
    for i in range( 1 , len(Att1[ index ]) , 1 ):
        if Att1[ index ][i] == x[ index ][j]:
            y[ index ][j] = y[ index ][j] + 1
        else:
            x[ index ].append( Att1[ index ][i] )
            y[ index ].append( 1 )
            j = j + 1
```
