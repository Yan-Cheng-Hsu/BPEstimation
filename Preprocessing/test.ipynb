```python
AttributeTable['R-peakIndex'] = []
AttributeTable['R-peakValue'] = []
def get_peakIndex_and_peakValue_of_ecg( startIndex , endIndex ):
    for i in range( startIndex , endIndex + 1 , 1 ):
        #get peak index of ecg
        temp = []
        k = 0;
        while k < len(RearrangedData['ecg'][i]):
            working_data, measures = hp.process(RearrangedData['ecg'][i][k:( k+250 )], 125.0)
            
            #shiftedIndex is for index shifting
            shiftedPeakIndex = [x+k for x in working_data['peaklist']]
            
            
            #find the rolling mean of hr
            hr = []
            hr.append( 60 )
            for index in range( 0 , len(shiftedPeakIndex) - 1 , 1 ):
                hr.append( shiftedPeakIndex[index+1] - shiftedPeakIndex[index] )
            #rolling( window_size, min_periods ) 
            RollingMeanofHR = pd.Series(hr).rolling( 5 , min_periods=1 ).mean()
            
            #check if there is abnormal index
            
            for y in range( 1 , len(hr) , 1 ):
                if hr[y] > ( RollingMeanofHR[y]*1.5 ):
                #there might be omissive peaks
                    searchingIndex = shiftedPeakIndex[y-1]
                    while ( shiftedPeakIndex[y] - searchingIndex ) > RollingMeanofHR[y]:
                        
                        local_max_index = searchingIndex
                        for d in range( searchingIndex , searchingIndex + RollingMeanofHR[y] , 1 ):
                            if RearrangedData['ecg'][i][d] > RearrangedData['ecg'][i][local_max_index]:
                                local_max_index = d
                        
                        shiftedPeakIndex.append(local_max_index)
                        searchingIndex = searchingIndex + RollingMeanofHR[y]

                if hr[y] < ( RollingMeanofHR[y]/1.5 ):
                    if RearrangedData['ecg'][i][shiftedPeakIndex[y]] > RearrangedData['ecg'][i][shiftedPeakIndex[y-1]]:
                        del shiftedPeakIndex[y-1]
                    else:
                        del shiftedPeakIndex[y]
            # TT is for DataType Transformation
            #since datatype of temp is list but working_data['peaklist'] is a numpy.ndarray.
            TT = list(shiftedPeakIndex)
            temp = temp + TT
            k = k + 250
            
        #temp = np.sort( temp )
        temp = np.array( temp )
        temp.sort();
        AttributeTable['R-peakIndex'].append( temp )
        
        #get peak value of ecg
        temp2 = []
        temp2 = np.array( temp2 )
        for j in range( len(AttributeTable['R-peakIndex'][i]) ):
            temp2 = np.append( temp2 , RearrangedData['ecg'][i][AttributeTable['R-peakIndex'][i][j]] ) 
        AttributeTable['R-peakValue'].append(temp2)
    return


get_peakIndex_and_peakValue_of_ecg(0,0)



#working_data, measures = hp.process(RearrangedData['ecg'][0], 125.0)
#hp.plotter(working_data, measures)
#working_data
```
